```python
#hyperparameters and evaluating using cross-validation. 
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# Define the hyperparameters to search through
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

best_score = 0
best_params = None

# Iterate through all parameter combinations
for max_depth in param_grid['max_depth']:
    for min_samples_split in param_grid['min_samples_split']:
        for min_samples_leaf in param_grid['min_samples_leaf']:
            # Create and evaluate the model with current parameters using cross-validation
            model = DecisionTreeClassifier(max_depth=max_depth,
                                           min_samples_split=min_samples_split,
                                           min_samples_leaf=min_samples_leaf,
                                           random_state=42)
            
            scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
            avg_score = scores.mean()
            
            # Check if current parameters yield a better score
            if avg_score > best_score:
                best_score = avg_score
                best_params = {
                    'max_depth': max_depth,
                    'min_samples_split': min_samples_split,
                    'min_samples_leaf': min_samples_leaf
                }

# Output the best parameters found
print("Best Hyperparameters:", best_params)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Create a DecisionTreeClassifier with the best hyperparameters
best_dt_classifier = DecisionTreeClassifier(
    max_depth=best_params['max_depth'],
    min_samples_split=best_params['min_samples_split'],
    min_samples_leaf=best_params['min_samples_leaf'],
    random_state=42
)

# Train the model with the best hyperparameters on the full training set
best_dt_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred_best_dt = best_dt_classifier.predict(X_test)

```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
X = data.drop('y', axis=1)  # Features
y = data['y']  # Target variable

# Encode categorical variables
X = pd.get_dummies(X, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'poutcome'])

# Convert 'yes' and 'no' labels to 1 and 0
y_binary = y.map({'yes': 1, 'no': 0})

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Initialize Decision Tree Classifier without Hyperparameter Tuning
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

# Predictions and evaluation metrics for Decision Tree before tuning
y_pred_dt = dt_classifier.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)

# Define the hyperparameters to search through
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Evaluate hyperparameters using cross-validation
best_score = 0
best_params = None

for max_depth in param_grid['max_depth']:
    for min_samples_split in param_grid['min_samples_split']:
        for min_samples_leaf in param_grid['min_samples_leaf']:
            model = DecisionTreeClassifier(max_depth=max_depth,
                                           min_samples_split=min_samples_split,
                                           min_samples_leaf=min_samples_leaf,
                                           random_state=42)
            scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
            avg_score = scores.mean()
            
            if avg_score > best_score:
                best_score = avg_score
                best_params = {
                    'max_depth': max_depth,
                    'min_samples_split': min_samples_split,
                    'min_samples_leaf': min_samples_leaf
                }

# Create a DecisionTreeClassifier with the best hyperparameters
best_dt_classifier = DecisionTreeClassifier(
    max_depth=best_params['max_depth'],
    min_samples_split=best_params['min_samples_split'],
    min_samples_leaf=best_params['min_samples_leaf'],
    random_state=42
)

# Train the model with the best hyperparameters
best_dt_classifier.fit(X_train, y_train)

# Predictions and evaluation metrics for Decision Tree after tuning
y_pred_best_dt = best_dt_classifier.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best_dt)
precision_best = precision_score(y_test, y_pred_best_dt)
recall_best = recall_score(y_test, y_pred_best_dt)
f1_best = f1_score(y_test, y_pred_best_dt)

# Store evaluation metrics in a DataFrame
metrics = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Decision Tree Before Tuning': [accuracy_dt, precision_dt, recall_dt, f1_dt],
    'Decision Tree After Tuning': [accuracy_best, precision_best, recall_best, f1_best]
})

# Print the comparison of evaluation metrics
print("Comparison of Evaluation Metrics:")
print(metrics)

# Plotting the metrics for comparison
metrics_melted = metrics.melt(id_vars='Metric', var_name='Classifier', value_name='Score')

plt.figure(figsize=(10, 6))
sns.barplot(x='Metric', y='Score', hue='Classifier', data=metrics_melted)
plt.title('Comparison of Evaluation Metrics')
plt.xlabel('Metric')
plt.ylabel('Score')
plt.legend(title='Classifier')
plt.show()

# Display confusion matrix for best Decision Tree after tuning
conf_matrix_best = confusion_matrix(y_test, y_pred_best_dt)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_best, annot=True, fmt='d', cmap='YlGn',
            xticklabels=['Predicted No', 'Predicted Yes'],
            yticklabels=['Actual No', 'Actual Yes'])
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix for Best Decision Tree Classifier')
plt.show()
```
